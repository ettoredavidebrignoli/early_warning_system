## Categorization for feature Selection

We now focus on organizing and transforming the financial data. In particular, we CATEGORIZE our features (time series of financial instruments) based on their **asset class**(like Equities, Commodities, and Fixed Income) and **geographical region**. Then, after applying the transformations necessary in order to make the data stationary, we calculate and visualize the correlation between the feature using a heatmap, helping to understand the relationships and dependencies within the data.

```python
# List of all the features
features = X_df.columns.tolist()

# Categorize by ASSET CLASS
features_by_assetclass = {
    # Equities (e.g., stock indices and related instruments)
    'Equities': ['MXUS', 'MXEU', 'MXJP', 'MXCN', 'MXIN', 'MXRU', 'MXBR','VIX'],

    # Commodities (e.g., raw materials, energy, precious metals)
    'Commodities': ['XAUBGNL', 'CRY', 'Cl1', 'BDIY'],

    # Fixed Income (e.g., government bonds, interest rates, credit spreads)
    'Fixed_income': ['EONIA', 'GTDEM10Y', 'GTDEM2Y', 'GTDEM30Y',
    'GTGBP20Y', 'GTGBP2Y', 'GTGBP30Y','GTITL10YR', 'GTITL2YR', 'GTITL30YR',
    'GTJPY10YR', 'GTJPY2YR', 'GTJPY30YR', 'US0001M', 'USGG2YR', 'USGG30YR', 'USGG3M','GT10'],

    # Currencies (e.g., foreign exchange and currency-related instruments)
    'Currencies': ['DXY', 'GBP', 'JPY'],

    # Credit Instruments (e.g., corporate bonds, credit spreads)
    'Credit Instruments': ['LF94TRUU', 'LF98TRUU', 'LG30TRUU', 'LMBITR', 'LP01TREU',
                           'LUACTRUU', 'LUMSTRUU', 'EMUSTRUU'],

    # Economic Indicators (e.g., macroeconomic indicators)
    'Economic Indicators': ['ECSURPUS']
}

# Categorize by REGION
features_by_region = {
    'US': ['MXUS', 'GT10','LG30TRUU','USGG2YR', 'USGG30YR', 'USGG3M', 'LF98TRUU', 'LUACTRUU', 'LUMSTRUU', 'US0001M', 'VIX', 'ECSURPUS'],
    'Europe/UK': ['GTGBP2Y', 'GTGBP20Y', 'GTGBP30Y', 'GBP','MXEU', 'GTDEM10Y', 'GTDEM2Y', 'GTDEM30Y', 'GTITL10YR', 'GTITL2YR', 'GTITL30YR', 'EMUSTRUU', 'EONIA', 'LP01TREU'],
    'Asia': ['MXJP', 'GTJPY10YR', 'GTJPY2YR', 'GTJPY30YR', 'JPY', 'MXCN', 'MXIN'],
    'Global/Other': ['MXBR', 'MXRU', 'BDIY', 'CRY', 'XAUBGNL', 'LF94TRUU','LMBITR']
}

# Function to get features by category
def get_features_by_category(category_name):
    return features_by_assetclass.get(category_name, [])

# Function to get features by region
def get_features_by_region(region_name):
    return features_by_region.get(region_name, [])

# Now, we create a dictionary for storing processed features
processed_features = {}


########################## STATIONARITY ############################

# Now we apply the transformations necessary to make the time series stationary, according to the asst class
# that we are considering. In fact, models based on statistical distributions assume that the data is stationary, i.e., its statistical properties (mean, variance)
# do not change over time. This is particularly important for Anomaly Detection models, which rely on estimating a stable distribution.

for category, category_features in features_by_assetclass.items():
    if category == 'Equities' or category == 'Commodities':
        # Apply log-differences for Equities and Commodities (positive values only)
        for feature in category_features:
            if feature in X_df.columns:
                processed_features[feature] = np.diff(np.log(X_df[feature]))  # Log-difference
    elif category == 'Fixed Income':
        # Apply first differences for Fixed Income (can be negative)
        for feature in category_features:
            if feature in X_df.columns:
                processed_features[feature] = np.diff(X_df[feature])  # First-difference
    else:
        # Directly assign the feature if no transformation is needed
        for feature in category_features:
            if feature in X_df.columns:
                processed_features[feature] = X_df[feature]

# Convert processed features back to a DataFrame for further analysis
processed_df = pd.DataFrame(processed_features, index=X_df.index[1:])  # Adjust for differences
processed_df.head()  # Check the first few rows

############################# CORRELATION ###############################
# Calculate the correlation matrix
correlation_matrix = processed_df.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(20, 15))  # Adjust the figure size if needed
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix of Assets')
plt.show()
```

# MVG model with classes of features



We can now explore the effectiveness of the **baseline MVG mode**l by applying it not only to the entire set of financial features, but also to **subsets of features**, according to the previous categorizations. This segmented approach aims to understand if anomalies are better detected within specific market segments or globally, and to evaluate the model's performance (measured by metrics like Precision, Recall, and F1 Score) in these different financial contexts. Essentially, we are trying to build a simple statistical detector to flag moments when the behavior of a specific set of financial instruments deviates significantly from their typical patterns, potentially indicating a shift towards a risk-on or risk-off regime.

## Baseline MVG model on US features



```python
from scipy.stats import multivariate_normal
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.utils import shuffle
from sklearn.preprocessing import StandardScaler

US_features = features_by_region['US']
X_US = processed_df[US_features].dropna().values
y_US = y_stationary

# Step 1: Creating training/cross-validation/test set with reshuffling
# Reshuffle the data (this will break down autocorrelation)
X_shuffled, y_shuffled = shuffle(X_US, y_US, random_state=42)

# Separate normal and anomalous examples
X_normal = X_shuffled[y_shuffled == 0]
X_anomaly = X_shuffled[y_shuffled == 1]

# Calculate sizes for each set
n_normal = X_normal.shape[0]
n_anomaly = X_anomaly.shape[0]

# Training set: 80% of normal examples
train_size = int(0.8 * n_normal)
X_train = X_normal[:train_size]

# Cross-validation set: 10% of normal examples and 50% of anomalies
cv_normal_size = int(0.1 * n_normal)
cv_anomaly_size = int(0.5 * n_anomaly)
X_cv_normal = X_normal[train_size:train_size + cv_normal_size]
X_cv_anomaly = X_anomaly[:cv_anomaly_size]
X_cross_val = np.vstack((X_cv_normal, X_cv_anomaly))
y_cross_val = np.hstack((np.zeros(cv_normal_size), np.ones(cv_anomaly_size)))

# Test set: 10% of normal examples and 50% of anomalies
X_test_normal = X_normal[train_size + cv_normal_size:]
X_test_anomaly = X_anomaly[cv_anomaly_size:]
X_test = np.vstack((X_test_normal, X_test_anomaly))
y_test = np.hstack((np.zeros(len(X_test_normal)), np.ones(len(X_test_anomaly))))

# We standardize the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_cross_val = scaler.transform(X_cross_val)
X_test = scaler.transform(X_test)

print(f"Training set size: {X_train.shape[0]} (all normal)")
print(f"Cross-validation set size: {X_cross_val.shape[0]} ({cv_normal_size} normal, {cv_anomaly_size} anomalies)")
print(f"Test set size: {X_test.shape[0]} ({len(X_test_normal)} normal, {len(X_test_anomaly)} anomalies)")

# Step 2: Training the model (estimating parameters of multivariate Gaussian)
# Calculate mean vector
mu = np.mean(X_train, axis=0)

# Calculate covariance matrix
sigma = np.cov(X_train, rowvar=False)

print(f"Mean vector shape: {mu.shape}")
print(f"Covariance matrix shape: {sigma.shape}")

# Step 3: Fine tuning the hyperparameter, the threshold ϵ
# Function to calculate multivariate Gaussian PDF
def multivariate_gaussian_pdf(X, mu, sigma):
    """Calculate the multivariate Gaussian probability density function"""
    n = mu.shape[0]

    # Handle potential numerical issues with the covariance matrix
    # Add a small regularization term to ensure positive definiteness
    sigma_reg = sigma + np.eye(n) * 1e-8

    # Calculate determinant and inverse
    try:
        det = np.linalg.det(sigma_reg)
        inv = np.linalg.inv(sigma_reg)
    except np.linalg.LinAlgError:
        # If still having issues, use pseudo-inverse
        print("Warning: Using pseudo-inverse for covariance matrix")
        det = max(np.linalg.det(sigma_reg), 1e-10)
        inv = np.linalg.pinv(sigma_reg)

    # Calculate PDF for each example
    p = np.zeros(X.shape[0])
    for i in range(X.shape[0]):
        x_mu = X[i] - mu
        p[i] = (1.0 / (np.power(2 * np.pi, n/2) * np.sqrt(det))) * \
               np.exp(-0.5 * x_mu.dot(inv).dot(x_mu))

    return p

# Compute the probability density function for the cross-validation set
p_cv = multivariate_gaussian_pdf(X_cross_val, mu, sigma)

# Find the range of epsilon values to search
min_epsilon = np.min(p_cv)
max_epsilon = np.max(p_cv)
step_size = (max_epsilon - min_epsilon) / 1000

# Find the best epsilon using F1 score
best_epsilon = 0
best_f1 = 0
best_precision = 0
best_recall = 0

epsilon_values = np.arange(min_epsilon, max_epsilon, step_size)
f1_scores = []
precisions = []
recalls = []

for epsilon in epsilon_values:
    predictions = (p_cv < epsilon).astype(int)

    # Calculate metrics
    precision = precision_score(y_cross_val, predictions, zero_division=0)
    recall = recall_score(y_cross_val, predictions, zero_division=0)

    # Calculate F1 score
    if precision + recall > 0:  # Avoid division by zero
        f1 = 2 * precision * recall / (precision + recall)
    else:
        f1 = 0

    f1_scores.append(f1)
    precisions.append(precision)
    recalls.append(recall)

    if f1 > best_f1:
        best_f1 = f1
        best_epsilon = epsilon
        best_precision = precision
        best_recall = recall

print(f"Best F1 score on CV set: {best_f1:.4f}")
print(f"Best Epsilon: {best_epsilon:.8e}")
print(f"Corresponding Precision: {best_precision:.4f}")
print(f"Corresponding Recall: {best_recall:.4f}")

from sklearn.metrics import confusion_matrix

# Step 4: Testing the model

# Compute the probability density function for the test set
p_test = multivariate_gaussian_pdf(X_test, mu, sigma)

# Make predictions using the best epsilon
predictions = (p_test < best_epsilon).astype(int)

# Calculate metrics
MVG_precision_US = precision_score(y_test, predictions, zero_division=0)
MVG_recall_US = recall_score(y_test, predictions, zero_division=0)
MVG_f1_US = f1_score(y_test, predictions, zero_division=0)

print("\nTest set performance:")
print(f"Precision US: {MVG_precision_US:.4f}")
print(f"Recall US: {MVG_recall_US:.4f}")
print(f"F1 Score US: {MVG_f1_US:.4f}")
```

# PCA per asset class cluster

The next section focuses on applying **Principal Component Analysis (PCA)** to different groups of financial features based on their asset class. This helps summarizing the main underlying movements and relationships within each group, reducing many potentially correlated time series into a smaller number of principal components (in this case just the first component for each group). This creates a new set of features, where each feature represents a **summary of the collective behavior of assets within a specific class**.
The components are then used to train and evaluate a Multivariate Gaussian (MVG) anomaly detection model.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Function: Apply PCA to a given list of features
def apply_pca(df, feature_list, prefix, n_components=1):
    df_subset = df[feature_list].dropna()  # Drop rows with NaNs
    if df_subset.shape[0] < 2:
        return pd.DataFrame(index=df.index, columns=[f'{prefix}_pca1'])  # Return NaNs if not enough data
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(df_subset)
    pca = PCA(n_components=n_components)
    X_pca = pca.fit_transform(X_scaled)
    pca_df = pd.DataFrame(X_pca, index=df_subset.index, columns=[f'{prefix}_pca{i+1}' for i in range(n_components)])

    # Reindex to match original DataFrame shape
    pca_df = pca_df.reindex(df.index)
    return pca_df

# Apply PCA to each category and add the result to the processed_df

processed_df_pca = processed_df.copy()
for category, feature_list in features_by_assetclass.items():
    valid_features = [f for f in feature_list if f in processed_df_pca.columns]
    if len(valid_features) >= 2:
        pca_result = apply_pca(processed_df_pca, valid_features, prefix=category.lower())
        processed_df_pca = pd.concat([processed_df_pca, pca_result], axis=1)

# Final list of PCA features
pca_columns = [col for col in processed_df_pca.columns if 'pca' in col]
print(f"Added PCA components: {pca_columns}")

# Select only PCA features
X_pca_only = processed_df_pca[[col for col in processed_df_pca.columns if 'pca' in col]].copy()

# Align y_stationary with X_pca
temp_df = pd.DataFrame({'y': y_stationary}, index=processed_df_pca.index)  # Adjust for differences
aligned_df = pd.concat([X_pca_only, temp_df], axis=1)

# Now drop NaNs from the aligned DataFrame
aligned_df = aligned_df.dropna()

# Separate X and y
X = aligned_df[[col for col in aligned_df.columns if 'pca' in col]].values
y = aligned_df['y'].values

###### MVG ######
from sklearn.utils import shuffle
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import precision_score, recall_score, f1_score

# Shuffle
X_shuffled, y_shuffled = shuffle(X, y, random_state=42)

# Separate normal/anomalous
X_normal = X_shuffled[y_shuffled == 0]
X_anomaly = X_shuffled[y_shuffled == 1]

# Train/CV/Test split
train_size = int(0.8 * len(X_normal))
cv_normal_size = int(0.1 * len(X_normal))
cv_anomaly_size = int(0.5 * len(X_anomaly))

X_train = X_normal[:train_size]
X_cv = np.vstack((X_normal[train_size:train_size+cv_normal_size], X_anomaly[:cv_anomaly_size]))
y_cv = np.hstack((np.zeros(cv_normal_size), np.ones(cv_anomaly_size)))

X_test = np.vstack((X_normal[train_size+cv_normal_size:], X_anomaly[cv_anomaly_size:]))
y_test = np.hstack((np.zeros(len(X_normal) - train_size - cv_normal_size), np.ones(len(X_anomaly) - cv_anomaly_size)))

# Standardization
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_cv = scaler.transform(X_cv)
X_test = scaler.transform(X_test)

# ------------------------
# MVG model: train, tune epsilon, evaluate
# ------------------------

# Estimate parameters
mu = np.mean(X_train, axis=0)
sigma = np.cov(X_train, rowvar=False)

# Tune epsilon
p_cv = multivariate_gaussian_pdf(X_cv, mu, sigma)
epsilon_range = np.linspace(p_cv.min(), p_cv.max(), 1000)

best_f1, best_epsilon, best_precision, best_recall = 0, 0, 0, 0
for epsilon in epsilon_range:
    preds = (p_cv < epsilon).astype(int)
    precision = precision_score(y_cv, preds, zero_division=0)
    recall = recall_score(y_cv, preds, zero_division=0)
    if precision + recall > 0:
        f1 = 2 * precision * recall / (precision + recall)
        if f1 > best_f1:
            best_f1 = f1
            best_epsilon = epsilon
            best_precision = precision
            best_recall = recall

# Evaluate on test set
p_test = multivariate_gaussian_pdf(X_test, mu, sigma)
predictions = (p_test < best_epsilon).astype(int)

print("\n Test Set Performance:")
print(f"Precision: {precision_score(y_test, predictions, zero_division=0):.4f}")
print(f"Recall: {recall_score(y_test, predictions, zero_division=0):.4f}")
print(f"F1 Score: {f1_score(y_test, predictions, zero_division=0):.4f}")
```

# Univariate correlation and feature importance




We now want to identify which individual features in our dataset are most strongly related to the 'risk-on/risk-off' anomaly status, and what is their contribution. We do it by calculating:

*  The **Pearson correlation** coefficient between each feature and the
target variable.
* The contribution of each individual financial feature to the **Mahalanobis distance** for the anomalous data points. The Mahalanobis distance is a measure of how far a point is from the mean of a distribution, taking into account the correlation and variance of the data. In the context of anomaly detection, a large Mahalanobis distance can indicate an unusual observation.

Pearson correlation


```python
y = pd.Series(y_stationary, index=processed_df.index, name='target')
corrs = processed_df.apply(lambda col: col.corr(y))

# Select the 7 features with the highest |correlation| values
top7_corr = corrs.abs().sort_values(ascending=False).head(7).index.tolist()


print("Top 7 features by univariate correlation (absolute value):")
print(top7_corr)
```

Contributions to the Mahalanobis distance


```python
# Compute mean vector and covariance matrix
X = processed_df.values
mu = np.mean(X, axis=0)
Sigma = np.cov(X, rowvar=False)

inv_Sigma = np.linalg.inv(Sigma)

# Select only the anomalous observations (target == 1)
anom_idx = y[y == 1].index
X_anom = processed_df.loc[anom_idx].values

diff = X_anom - mu  # shape (n_anom, n_feat)
left = diff @ inv_Sigma  # shape (n_anom, n_feat)
contrib = left * diff    # n_anomalies x n_features (contribution of each feature to each anomaly)

# Mean absolute importance per feature
feat_importance = np.mean(np.abs(contrib), axis=0)
feat_names = processed_df.columns.tolist()

# Take the top-7 features
top_mahala_indices = np.argsort(feat_importance)[::-1][:7]
top_mahala_features = [feat_names[i] for i in top_mahala_indices]

print(f"Top 7 features by Mahalanobis contribution: {top_mahala_features}")
```

# Engineered features


This section takes the previously processed financial data and engineers **five new features** by **combining existing time series**, aiming to capture more complex relationships within the market. From a financial perspective, these engineered features represent aggregated or relative measures of market behavior across different asset classes and regions.

These engineered features are then used to train and evaluate a **Multivariate Gaussian (MVG) anomaly detection model**, in addition to the features that we found to be the most correlated with the target variable and the most important according to Pearson importance.

In particular, we created five new features:
* **f1 = BDIY over_VIX**: Global Economic Activity vs. Market Fear. The BDIY is a measure of the cost of shipping raw materials globally, serving as a proxy for worldwide economic activity and demand for goods. The VIX is a measure of expected stock market volatility. The ratio of BDIY to VIX can be interpreted as a gauge of risk sentiment: High value suggests strong trade and low fear (risk-on). Low value suggests weak trade or high fear (risk-off).

* **f2 = LF98TRUU + LP01TREU**: Aggregated Credit Market Sentiment. Represents overall health or perceived risk in certain credit segments (e.g., corporate bonds, high-yield debt).Higher values may indicate better sentiment, lower values worse.

* **f3 = all yields product**: Product of 10-year (and 20-year for GBP) government bond yields from Germany, the UK, Italy, and Japan. Government bond yields reflect a combination of factors, including inflation expectations, economic growth outlook, and the perceived creditworthiness (sovereign risk) of the issuing country. Multiplying yields from major economies provides a consolidated view

* **f4= fx_prod**: This feature captures the combined movements of several major global currencies relative to a base (for DXY) or each other. It can indicate significant shifts or instability across key global currencies, and risk-on/risk-off periods are frequently accompanied by distinct movements in major currencies

* **f5 = MX_agg**: Aggregated Global Equity Performance. Represents the combined performance of major stock markets, directly indicating risk-on (high values) or risk-off (low values) sentiment. Stock markets are often seen as key indicators of risk sentiment and economic health.High values (or large positive changes) generally correspond to a risk-on environment, reflecting investor optimism and appetite for equity exposure.



Now we also use the features to train and evaluate an **XGBoost** classification model, which we found to be the model giving the best performances for our problem.

```python
from sklearn.utils import shuffle
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import precision_score, recall_score, f1_score
import numpy as np
from scipy.stats import multivariate_normal


# 1. Construction of  5 NEW COMPOSED FEATURES
df_feats = pd.DataFrame(index=processed_df.index)

numerator = processed_df['BDIY'].values
denominator = processed_df['VIX'].values
df_feats['f1_BDIY_over_VIX'] = np.where(denominator != 0, numerator / denominator, np.nan) # we replace division by zero results with NaN

df_feats['f2_LF98TRUU_plus_LP01TREU'] = (processed_df['LF98TRUU'] + processed_df['LP01TREU'] )
df_feats['f3_all_yields_prod'] = (processed_df['GTDEM10Y']* processed_df['GTGBP20Y']* processed_df['GTITL10YR']* processed_df['GTJPY10YR'])
df_feats['f4_fx_prod'] = (processed_df['DXY'] * processed_df['GBP'] * processed_df['JPY'])
df_feats['f5_MX_agg'] = (processed_df['MXUS'] + processed_df['MXEU'] + processed_df['MXCN'] + processed_df['MXJP']
)

df_feats = df_feats.dropna()

# Now df_feats is the DataFrame with the 5 composite features

# 2. Identify Already Used Features in Engineered Features
# We now take the list of original features that were used to create the engineered features
engineered_feature_components = [
    'BDIY', 'VIX', 'LF98TRUU', 'LP01TREU', 'GTDEM10Y', 'GTGBP20Y',
    'GTITL10YR', 'GTJPY10YR', 'DXY', 'GBP', 'JPY', 'MXUS', 'MXEU', 'MXCN', 'MXJP'
]

# 3. Select Additional Features: we want to add to this dataframe the features from
#    top_corr_features or top_mahala_features that were NOT used for computing the engineered ones

additional_features_to_add = []

# Check features from top_corr_features
for feature in top7_corr:
    if feature not in engineered_feature_components and feature not in additional_features_to_add:
        additional_features_to_add.append(feature)

# Check features from top_mahala_features
for feature in top_mahala_features:
     # Also ensure the feature exists in the original processed_df
    if feature not in engineered_feature_components and feature not in additional_features_to_add and feature in processed_df.columns:
        additional_features_to_add.append(feature)

print(f"\nAdditional features to add: {additional_features_to_add}")

# 4. Combine DataFrames
# Select the additional features from the original processed_df
if additional_features_to_add:
    df_additional = processed_df[additional_features_to_add]

    # Concatenate engineered features with additional features
    df_feats = pd.concat([df_feats, df_additional], axis=1)


############### MVG with the engineered features ###############

# We ensure that X (features) and y (target) are aligned and have the same length
y_aligned = pd.Series(y_stationary, index=processed_df.index, name='Y')
aligned_data = pd.concat([df_feats, y_aligned], axis=1).dropna()

# Separate features (X) and target (y) from the aligned DataFrame
X = aligned_data[df_feats.columns].values
y = aligned_data['Y'].values

print(f"Aligned and cleaned dataset - Shape X: {X.shape}, Shape y: {y.shape}")
print(f"Number of anomalies (Y=1) in the final dataset: {np.sum(y == 1)}")

# Dataset Split (Training, Cross-Validation, Test)

# Shuffle the data to break any temporal autocorrelation for splitting
X_shuffled, y_shuffled = shuffle(X, y, random_state=42)

# Separate normal examples (Y=0) from anomalous examples (Y=1)
X_normal = X_shuffled[y_shuffled == 0]
X_anomaly = X_shuffled[y_shuffled == 1]

# Dataset splitting: 80% normal for training, 10% normal + 50% anomalies for CV, 10% normal + 50% anomalies for Test
n_normal = X_normal.shape[0]
n_anomaly = X_anomaly.shape[0]

train_size = int(0.8 * n_normal)
cv_normal_size = int(0.1 * n_normal)
cv_anomaly_size = int(0.5 * n_anomaly)

# Ensure there is enough data in each category for the desired split
if train_size == 0 or cv_normal_size == 0 or cv_anomaly_size == 0 or \
   len(X_normal) - train_size - cv_normal_size == 0 or len(X_anomaly) - cv_anomaly_size == 0:
    print("Error: Not enough normal or anomalous examples to perform the desired split.")

else:
    # Build training, cross-validation, and test sets
    X_train = X_normal[:train_size]

    X_cv_normal = X_normal[train_size : train_size + cv_normal_size]
    X_cv_anomaly = X_anomaly[:cv_anomaly_size]
    X_cross_val = np.vstack((X_cv_normal, X_cv_anomaly))
    y_cross_val = np.hstack((np.zeros(cv_normal_size), np.ones(cv_anomaly_size)))

    X_test_normal = X_normal[train_size + cv_normal_size :]
    X_test_anomaly = X_anomaly[cv_anomaly_size :]
    X_test = np.vstack((X_test_normal, X_test_anomaly))
    y_test = np.hstack((np.zeros(len(X_test_normal)), np.ones(len(X_test_anomaly))))

    print("\nSet sizes after split:")
    print(f"Training (Normal): {X_train.shape[0]}")
    print(f"Cross-Validation: {X_cross_val.shape[0]} ({cv_normal_size} Normal, {cv_anomaly_size} Anomalies)")
    print(f"Test: {X_test.shape[0]} ({len(X_test_normal)} Normal, {len(X_test_anomaly)} Anomalies)")

    # Standardization
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_cross_val_scaled = scaler.transform(X_cross_val)
    X_test_scaled = scaler.transform(X_test)

    # MVG Model Training (Parameter Estimation)
    mu = np.mean(X_train_scaled, axis=0)
    sigma = np.cov(X_train_scaled, rowvar=False)

    # Hyperparameter Optimization (epsilon) on Cross-Validation Set
    p_cv = multivariate_gaussian_pdf(X_cross_val_scaled, mu, sigma)

    # Determine the range of epsilon values to explore
    min_epsilon = np.min(p_cv)
    max_epsilon = np.max(p_cv)
    epsilon_values = np.linspace(min_epsilon, max_epsilon, 1000)  # Try 1000 values

    best_epsilon = 0
    best_f1 = 0
    best_precision = 0
    best_recall = 0

    # Iterate over each epsilon value and compute metrics
    for epsilon in epsilon_values:

        predictions = (p_cv < epsilon).astype(int)

        # Compute Precision, Recall, and F1 Score
        precision = precision_score(y_cross_val, predictions, zero_division=0)  # zero_division=0 avoids warning if no positive predictions
        recall = recall_score(y_cross_val, predictions, zero_division=0)

        # Compute F1 Score
        if (precision + recall) > 0:  # Avoid division by zero
            f1 = 2 * precision * recall / (precision + recall)
        else:
            f1 = 0

        if f1 > best_f1:
            best_f1 = f1
            best_epsilon = epsilon
            best_precision = precision
            best_recall = recall

    print(f"   Best epsilon found: {best_epsilon:.8e}")
    print(f"   Best F1 Score (CV): {best_f1:.4f}")
    print(f"   Corresponding Precision (CV): {best_precision:.4f}")
    print(f"   Corresponding Recall (CV): {best_recall:.4f}")


    # Final Evaluation on Test Set
    p_test = multivariate_gaussian_pdf(X_test_scaled, mu, sigma)
    predictions_test = (p_test < best_epsilon).astype(int)

    # Compute final metrics on the test set
    final_precision = precision_score(y_test, predictions_test, zero_division=0)
    final_recall = recall_score(y_test, predictions_test, zero_division=0)
    final_f1 = f1_score(y_test, predictions_test, zero_division=0)

    print("\n  Final performance on the Test set (with Engineered Features):")
    print(f"   Precision: {final_precision:.4f}")
    print(f"   Recall: {final_recall:.4f}")
    print(f"   F1 Score: {final_f1:.4f}")

```
