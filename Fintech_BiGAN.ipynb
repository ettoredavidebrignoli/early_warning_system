```python
# Step 0: Functions and utilities for comparison of supervised and unsupervised Anomaly Detection methods

from sklearn.metrics import roc_curve, auc
from sklearn.ensemble import RandomForestClassifier, IsolationForest
from sklearn.svm import OneClassSVM, SVC
from sklearn.neighbors import LocalOutlierFactor

# Set visualization style
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [12, 8]

# We already have X_train, X_cross_val, y_cross_val, X_test, y_test from previous code cells
# We'll standardize the data for better performance with many algorithms
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_cv_scaled = scaler.transform(X_cross_val)
X_test_scaled = scaler.transform(X_test)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix

# Set random seeds for reproducibility
np.random.seed(42)
torch.manual_seed(42)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(42)


# Check if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Convert data to PyTorch tensors
X_train_tensor = torch.FloatTensor(X_train_scaled).to(device)
X_cv_tensor = torch.FloatTensor(X_cv_scaled).to(device)
X_test_tensor = torch.FloatTensor(X_test_scaled).to(device)
```

```python
batch_size = 64
train_dataset = TensorDataset(X_train_tensor, X_train_tensor)  # Input = Output for autoencoder
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
```

```python
# Encoder
class Encoder(nn.Module):
    def __init__(self, input_dim, encoding_dim=16):
        super().__init__()
        # Calculate layer sizes for a gradually decreasing architecture
        layer1_size = input_dim // 2
        layer2_size = layer1_size // 2

        # Ensure minimum size
        layer1_size = max(layer1_size, encoding_dim * 2)
        layer2_size = max(layer2_size, encoding_dim)

        self.net = nn.Sequential(
            nn.Linear(input_dim, layer1_size),
            nn.BatchNorm1d(layer1_size),
            nn.ReLU(True),
            nn.Dropout(0.2),
            nn.Linear(layer1_size, layer2_size),
            nn.BatchNorm1d(layer2_size),
            nn.ReLU(True),
            nn.Dropout(0.2),
            nn.Linear(layer2_size, encoding_dim),
            nn.BatchNorm1d(encoding_dim),
            nn.ReLU(True)
        )

    def forward(self, x):
        return self.net(x)

# Generator
class Generator(nn.Module):
    def __init__(self, output_dim, encoding_dim=16):
        super().__init__()

        # Calculate layer sizes for a gradually decreasing architecture
        layer1_size = output_dim // 2
        layer2_size = layer1_size // 2

        # Ensure minimum size
        layer1_size = max(layer1_size, encoding_dim * 2)
        layer2_size = max(layer2_size, encoding_dim)

        self.net = nn.Sequential(
            nn.Linear(encoding_dim, layer2_size),
            nn.BatchNorm1d(layer2_size),
            nn.ReLU(True),
            nn.Dropout(0.2),
            nn.Linear(layer2_size, layer1_size),
            nn.BatchNorm1d(layer1_size),
            nn.ReLU(True),
            nn.Dropout(0.2),
            nn.Linear(layer1_size, input_dim),
            nn.Tanh()
        )
        self.encoding_dim = encoding_dim

    def forward(self, z):
        return self.net(z)

# Discriminator
class Discriminator(nn.Module):
    def __init__(self, input_dim, encoding_dim=16):
        super().__init__()

        # Calculate layer sizes for a gradually decreasing architecture
        dim = input_dim + encoding_dim
        layer1_size = dim // 2
        layer2_size = layer1_size // 2

        self.net = nn.Sequential(
            nn.Linear(dim, layer1_size),
            nn.BatchNorm1d(layer1_size),
            nn.LeakyReLU(True),
            nn.Dropout(0.2),
            nn.Linear(layer1_size, layer2_size),
            nn.BatchNorm1d(layer2_size),
            nn.LeakyReLU(True),
            nn.Dropout(0.2),
            nn.Linear(layer2_size, 1),
            nn.Sigmoid()
        )

    def forward(self, x, z):
        input = torch.cat([x, z], dim=1)
        return self.net(input)

# Useful function
def sample_noise(batch_size, z_dim):
    return torch.randn(batch_size, z_dim)

# Loss
bce_loss = nn.BCELoss()



# Training Loop
def train_bigan(encoder, generator, discriminator, train_loader, num_epochs=100):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    encoder.to(device)
    generator.to(device)
    discriminator.to(device)

    opt_D = optim.Adam(discriminator.parameters(), lr=2e-4)
    opt_GE = optim.Adam(list(encoder.parameters()) + list(generator.parameters()), lr=2e-4)

    # For plotting
    d_losses = []
    ge_losses = []

    for epoch in range(num_epochs):
        for x_batch in train_loader:
            x = x_batch[0].to(device)
            batch_size = x.size(0)
            z = sample_noise(batch_size, generator.encoding_dim).to(device)

            # Train Discriminator
            z_enc = encoder(x)
            x_gen = generator(z)

            D_real = discriminator(x, z_enc)
            D_fake = discriminator(x_gen.detach(), z)

            loss_D = bce_loss(D_real, torch.ones_like(D_real)) + \
                     bce_loss(D_fake, torch.zeros_like(D_fake))

            opt_D.zero_grad()
            loss_D.backward()
            opt_D.step()

            # Train Generator + Encoder
            z_enc = encoder(x)
            x_gen = generator(z)

            D_real = discriminator(x, z_enc)
            D_fake = discriminator(x_gen, z)

            loss_GE = bce_loss(D_real, torch.zeros_like(D_real)) + \
                       bce_loss(D_fake, torch.ones_like(D_fake))

            opt_GE.zero_grad()
            loss_GE.backward()
            opt_GE.step()

        d_losses.append(loss_D.item())
        ge_losses.append(loss_GE.item())

        if (epoch + 1) % 10 == 0:
          print(f"Epoch [{epoch+1}/{num_epochs}], D Loss: {loss_D.item():.4f}, GE Loss: {loss_GE.item():.4f}")

    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(d_losses, label='Discriminator Loss')
    plt.plot(ge_losses, label='Generator + Encoder Loss')
    plt.xlabel('Epochs')
    plt.grid(True)
    plt.legend()
    plt.title('Training Losses')
    plt.show()


```

```python
input_dim = X_train_scaled.shape[1]
encoding_dim = min(16, input_dim // 4)
#encoding_dim = 20

encoder = Encoder(input_dim, encoding_dim).to(device)
generator = Generator(input_dim, encoding_dim).to(device)
discriminator = Discriminator(input_dim, encoding_dim).to(device)

train_bigan(encoder, generator, discriminator, train_loader, num_epochs=500)
```

```python
# Function to compute reconstruction error
def compute_reconstruction_error(generator, encoder, data_tensor):
    generator.eval()
    encoder.eval()
    with torch.no_grad():
        reconstructions = generator(encoder(data_tensor))
        # Compute MSE for each sample
        mse = ((reconstructions - data_tensor) ** 2).mean(dim=1).cpu().numpy()
    return mse
```

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
import seaborn as sns
import matplotlib.pyplot as plt
import torch

def test_bigan(x_test, y_test, encoder, generator, threshold, device=None):
    if device is None:
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # Assicura tensori torch
    x_tensor = torch.tensor(x_test, dtype=torch.float32)
    y_true = y_test if isinstance(y_test, torch.Tensor) else torch.tensor(y_test, dtype=torch.int)


    # Calcola gli anomaly score
    scores = compute_reconstruction_error(generator, encoder, x_tensor)

    # Classifica in base alla soglia
    preds = (scores > threshold).astype(int)

    # Confusion matrix
    cm = confusion_matrix(y_true.numpy(), preds)


    precision = precision_score(y_true, preds, zero_division=0)
    recall = recall_score(y_true, preds, zero_division=0)
    f1 = f1_score(y_true, preds, zero_division=0)

    print("\nBiGAN performance:")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")

    # Plot confusion matrix
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=["Normal", "Anomaly"], yticklabels=["Normal", "Anomaly"])
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title("Confusion Matrix")
    plt.tight_layout()
    plt.show()

    # ROC Curve
    fpr, tpr, _ = roc_curve(y_true.numpy(), scores)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (AUC = %0.4f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=1, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC)')
    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.show()

```
